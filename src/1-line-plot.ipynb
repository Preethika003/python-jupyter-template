### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plot
import numpy as np

try:
    # Generate x values from 0 to 10
    x = np.linspace(0, 10, 100)

    # Calculate y values as the square of x
    y = x**2

    # Create the plot
    plot.plot(x, y)

    # Setting the titles and labels
    plot.title("Line Plot of y = x^2")
    plot.xlabel("x values")
    plot.ylabel("y values (x^2)")

    # Display the plot
    plot.show()

except ImportError as e:
    print("Error: Module matplotlib and numpy maybe missing. Please install")
except Exception as e:
    print(f"An unexpected error has occurred: {e}")
```
