### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plot

try:
    # Create a 2x2 subplot layout
    fig, axis = plot.subplots(2, 2, figsize=(10, 8))

    # Data for the plots
    x = [1, 2, 3, 4, 5]
    y = [5, 4, 3, 2, 1]
    categories = ['A', 'B', 'C', 'D', 'E']
    values = [10, 15, 20, 25, 30]

    # Line chart in the first subplot
    axis[0, 0].plot(x, y)
    axis[0, 0].set_title('Line Chart')

    # Bar chart in the second subplot
    axis[0, 1].bar(categories, values)
    axis[0, 1].set_title('Bar Chart')

    # Scatter plot in the third subplot
    axis[1, 0].scatter(x, y)
    axis[1, 0].set_title('Scatter Plot')

    # Pie chart in the fourth subplot
    axis[1, 1].pie(values, labels=categories, autopct='%1.1f%%')
    axis[1, 1].set_title('Pie Chart')

    try:
        # Display the plot
        plot.tight_layout()
        plot.show()
    except Exception as plot_error:
        print(f"Error during plot generating: {plot_error}")    
    
except ImportError as e:
    print("Error: Module matplotlib and numpy maybe missing. Please install")
except Exception as e:
    print(f"An unexpected error has occurred: {e}")
```
