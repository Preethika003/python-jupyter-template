### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plot

# Data 
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create scatter plot
plot.scatter(x, y, color='blue')

# Annotate each point with its (x, y) value
for i in range(len(x)):
    plot.text(x[i]+.1, y[i], f'({x[i]}, {y[i]})', fontsize=10, ha='left', color='green')

# Setting the title and labels
plot.title('Scatter Plot with Annotations')
plot.xlabel('x values')
plot.ylabel('y values')

# Display plot
plot.show()
```
