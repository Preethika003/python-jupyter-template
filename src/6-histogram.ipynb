### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt

try:
    # Data
    histoData = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

    # Create histogram
    plt.hist(histoData, bins=range(1, 7), edgecolor='black', color='cyan')

    # Customize the plot
    plt.title('Plotting of Histogram')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

    # Show the plot
    plt.show()

except ImportError as e:
    print("Error: Module matplotlib and numpy maybe missing. Please install")
except Exception as e:
    print(f"An unexpected error has occurred: {e}")
```
