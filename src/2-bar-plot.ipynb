### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plot

try:
    # Data categories and values
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]

    # Creating a bar plot
    plot.bar(categories, values, color=['red', 'green', 'blue', 'orange'])

    # Setting the Title
    plot.title('Distrubution of Categories and Values')

    try:
        # Display the bar plot
        plot.show()
    except Exception as plot_error:
        print(f"Error during plot generating: {plot_error}")    

except ImportError as e:
    print("Error: Module matplotlib and numpy maybe missing. Please install")
except Exception as e:
    print(f"An unexpected error has occurred: {e}")
```
